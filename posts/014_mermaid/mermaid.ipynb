After learning about mermaid sequence in class and how information could be displayed in flowcharts, I decided to experiment with this method by narrating a different happy ever after ending for Ariel, The Little Mermaid. (Note my play on words). By playing with the sequence, I showed at a basic level how Ariel's fate could have turned out if there were different decisions made at the milestone where she fell in love with Prince Eric. Doing so is a bit silly to see as we know all Disney movies and princesses have their happy ending.

```{mermaid}
flowchart LR
  A[Ariel dreams of living on land] --> B(Ariel rescues Prince Eric)
  B --> C{Ariel falls in love with Eric}
  C --> D[Ariel and Eric marry]
  C --> E[Ariel leaves Eric]
```


Working with this a little more, we could demonstrate how a conversation between Ariel and Eric may go depending on the decisions at the end of the flow chart. If decision 1 happened and we see they got married, perhaps they would have had a conversation like this:

```{mermaid}
sequenceDiagram
    participant A as Ariel
    participant B as Eric

    A->>B: Prince Eric, I love you so much!
    B->>A: Oh Princess Ariel, will you marry me?
    A->>B: Yes Eric, I do!
    B->>A: This is my dream come true!
    A->>B: Should we have our wedding on land or under the sea?
```

But, if fate didn't end up this way. How would a conversation between the two have gone? Maybe something like this...

```{mermaid}
sequenceDiagram
    participant A as Ariel
    participant B as Eric
    participant C as Ursula

    A->>B: Eric, why won't you come back under the sea with me?
    B->>A: Princess Ariel, my life is on land.
    A->>B: We will never work then.
    B->>A: But Ariel, I love you! Don't leave!
    B->>C: The plan worked. She won't come on land again your highness.
```

Maybe the villian might win in this story. Hopefully not!
