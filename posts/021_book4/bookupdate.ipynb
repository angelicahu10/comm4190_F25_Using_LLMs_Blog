As Dr. Buomalmwini continued her research to expose the dicsrimination present in large corporation softwares, we see large giants either accept or defend this thesis. Companies such as IBM and Windows were open to these criticims and used them to better their practices. Amazon, however, attacked her accusations made in the "Audctionable Auditing" paper, ready to back their organization. In responding to Amazon, Buomlamwini lost much support and backing that she had prior to this publication. Not surprisingly, many corporations were scared of facing such a large influential parent company like Amazon.

What's interesting about the news is we often see history repeat itself. This sceario of power is similar to modern day trials regarding institutions and their complying to federal regulations under Trump's administration. While powerful corporations claim to stand for what is right, we see them succumb to larger giants in power in order to protect what they have. They become powerless or or unable to go against the status quo.

This was an interesting, yet saddening, observation to stumble upon and instead rather highlights all the more importance of bringing into light advocating for change. By bringing awareness to these stituations, the spotlight on those who have power and influence should be changed and corporations who are doing harm must be held accountable.
