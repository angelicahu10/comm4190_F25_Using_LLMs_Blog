This year, I entered my second decade of living, and with growing up and finding your sense of identity often comes many self-doubting crises. What should I study? Will I find a job? Do I like my friends? Where do I want to live when I'm older? Finding identity and sense of self isn't easy and something that often comes with time; and stress about a career is often normal when we start to consider internships and working in the real world. As we know, there are many applications that use AI to help tailor resumes, apply to job positions, etc. So, I started wondering, if I explained a brief explaination to an LLM, would they be able to help me figure out what I want to do?
