What is a magic 8 ball?

### ASSISTANT
A Magic 8 Ball is a novelty toy used for fortune-telling or seeking advice. It's a hollow plastic sphere, designed to look like an oversized black and white 8-ball from pool, that contains a 20-sided die floating in a dark liquid. Each face of the die has a different answer printed on it.

To use it, you ask the Magic 8 Ball a yes-or-no question, shake it, and then turn it over to reveal an answer that appears in a small window on the bottom. Answers range from affirmative ("It is certain," "Yes definitely"), to non-committal ("Reply hazy, try again," "Ask again later"), to negative ("Don't count on it," "My sources say no").

The Magic 8 Ball was invented in the 1940s and has become a popular cultural icon and toy that's still sold today.
