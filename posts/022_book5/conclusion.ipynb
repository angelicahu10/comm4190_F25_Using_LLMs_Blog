After reading the epilogue and finishing the novel Unmasking AI: My Mission To Protect What is Human In A World Of Machines by Dr. Joy Buolamwini, I decided to reflect on my experience as a reader with a final blog post. What this book essentially is, is an eloquent capture of Joy Buolamwini's journey as an academic at MIT to an industry leader in algorithmic justice in AI. From the first chapter to the last, we follow the intricate steps in how Buolamwini became so esteemed and intersested in algorithmic bias based on personal experiences. 

We start with the Aspire Mirror and the beginnings of facial classification technology which has become much more advanced in today's society. From there, Buolamwini takes her passion and fuels it into a fight for justice through raising awareness of the harms present in the coded gaze, specifically the effects on the "ex-coded." The book narrates the timeline of some of the works we found in a seperate blog post -- from her documentary, her TedTalk, her prescence at certain panels and conventions, and the Congressional hearing regarding AI with representatives such as AOC. It is interesting to see these events show up on the AJL website while reading about them unfolding from her perpsective in the novel. 

Buolamwini is a true inspiration, proving that if you are passionate about something or especially social change, all it takes is one individual to speak out and initiate a movement, drawing attention to issues at hand and collecting joint forces along the way. In committing to something, your voice becomes powerful and influential as an expert leader who is willing to be the change. Much work with algorithmic bias and how it could lead to greater, unintended harms, is in thanks to Dr. Joy Buolamwini's efforts and remains ever as important in today's modern world.
